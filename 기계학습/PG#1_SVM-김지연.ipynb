```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn as pd
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC

iris = datasets.load_iris()

X = iris["data"][:,(2,3)]
y = (iris["target"]==2).astype(np.float64)
```

```python
plt.figure()
plt.scatter(X[:,0], X[:,1], c=y)
plt.xlabel("petal length", fontsize=14)
plt.ylabel("petal width", fontsize=14)
plt.title("feature scatter", fontsize=18)
```

```python
# C 1 svm_clf_1

svm_clf_1 = Pipeline([
    ("scaler", StandardScaler()),
    ("linear_svc", LinearSVC(C=1, loss="hinge", max_iter=50000)),
])

svm_clf_1.fit(X,y)
print(svm_clf_1.named_steps["linear_svc"].coef_)
print(svm_clf_1.named_steps["linear_svc"].intercept_)

w= -svm_clf_1.named_steps["linear_svc"].coef_[0,0]/svm_clf_1.named_steps["linear_svc"].coef_[0,1]
b= -svm_clf_1.named_steps["linear_svc"].intercept_[0]/svm_clf_1.named_steps["linear_svc"].coef_[0,1]

line = svm_clf_1.named_steps["scaler"].inverse_transform([[-5,(-5*w)+b], [5,5*w+b]])

plt.figure()
plt.plot(line[:,0], line[:,1], "k:", label="LinearSVC")
plt.plot(X[:,0][y==1], X[:,1][y==1],'bo',label='lris-Virginica')
plt.plot(X[:,0][y==0], X[:,1][y==0],'yo',label='None-Virginica')
plt.plot(3.8,1.7,'r*')
print(svm_clf.predict([[3.8,1.7]]))

plt.xlabel("petal length", fontsize=14)
plt.ylabel("petal width", fontsize=14)
plt.legend(loc='lower right', fontsize=14)
plt.title("Linear SVM fitting", fontsize=18)
plt.axis([0,7,0,2.5])
plt.show()
```
